```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
#load data

def get_data_from_opengov_munich(resource_id, limit = 8000):
    params = params={
        'resource_id': resource_id, 
        'limit': limit ,
    }
    url = 'https://www.opengov-muenchen.de/api/action/datastore_search'
    r = requests.get(url, params=params).json()
    return pd.DataFrame(r['result']['records'])



df = get_data_from_opengov_munich("56a0c3c8-c2ea-4b42-bbd2-21cb72d80803", 80000)

```
